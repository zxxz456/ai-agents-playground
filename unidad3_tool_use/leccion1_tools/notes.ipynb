## ¿Qué es el Uso de Herramientas?

**Tool use** (también conocido como *function calling*) es la capacidad de extender las funcionalidades de Claude mediante la definición e invocación de **herramientas o funciones externas**.

Podemos darle a Claude acceso a un conjunto de herramientas predefinidas que puede invocar en cualquier momento. Estas herramientas nos permiten escribir código que realiza **tareas específicas o cómputos** que Claude no podría hacer por sí solo.

> **En resumen:** Tool use es un mecanismo para **potenciar** las capacidades de Claude conectándolo con el mundo exterior.

### Conceptos Clave

- **Herramienta (Tool):** Una función o servicio externo que Claude puede solicitar ejecutar
- **Esquema de entrada (Input Schema):** La definición formal de los parámetros que acepta una herramienta
- **Resultado de herramienta (Tool Result):** La respuesta que tu código devuelve a Claude después de ejecutar la herramienta
- **Function Calling:** Nombre alternativo para tool use, usado en otros modelos como GPT

## ¿Por Qué es Importante el Tool Use?

El uso de herramientas es una funcionalidad **revolucionaria** que desbloquea el verdadero potencial de los LLMs y abre un mundo de posibilidades. Es un componente crítico que puede mejorar significativamente el valor e impacto de tus aplicaciones.

| Beneficio | Descripción |
|-----------|-------------|
| **Extender capacidades** | Permite a Claude realizar tareas que **no podría hacer** con su funcionalidad integrada: cálculos en tiempo real, acceso a datos actualizados, operaciones sobre sistemas externos. |
| **Integrar con sistemas existentes** | Conecta a Claude con tu infraestructura actual: bases de datos, APIs internas, CRMs, ERPs y cualquier servicio backend. |
| **Automatizar tareas complejas** | Encadena múltiples herramientas para crear **flujos de trabajo automatizados** que involucren varios pasos o cómputos especializados. |
| **Mejorar la experiencia de usuario** | Entrega respuestas más **precisas, contextuales y personalizadas**. Los usuarios interactúan en lenguaje natural y Claude ejecuta las acciones apropiadas. |
| **Escalar y personalizar** | Agrega nuevas herramientas o modifica las existentes conforme evolucionen las necesidades. Máxima **flexibilidad** para iterar rápidamente. |

## ¿Cómo Funciona el Tool Use?

Lo más importante que debes entender es que **Claude NO ejecuta ningún código por sí mismo**. Nosotros le informamos sobre un conjunto de herramientas que puede *pedirnos* que ejecutemos, y luego es **nuestra responsabilidad** ejecutar el código subyacente y comunicarle los resultados.

> **Nota Clave:** Claude **no tiene acceso** a herramientas integradas del lado del servidor. Todas las herramientas deben ser **explícitamente proporcionadas** por ti en cada solicitud a la API.

### Esto significa que **TU** controlas:

- La **definición** de las herramientas disponibles (nombres, descripciones, esquemas)
- La **implementación** y ejecución de la lógica de cada herramienta
- La **seguridad** y validación de entradas/salidas
- El **flujo** de datos entre Claude y tus sistemas

### Los 4 Pasos del Tool Use

El uso de herramientas con Claude sigue un flujo bien definido:

```
 PASO 1                    PASO 2                   PASO 3                   PASO 4
 (API Request)             (API Response)           (API Request)            (API Response)
 ─────────────             ──────────────           ─────────────            ──────────────

 ┌───────────┐            ┌───────────────┐        ┌──────────────┐        ┌───────────────┐
 │ Enviar a  │            │ Claude decide │        │ Extraer tool │        │ Claude usa    │
 │ Claude:   │───────────►│ usar tool y   │───────►│ input, correr│───────►│ resultado para│
 │           │            │ genera request│        │ código real, │        │ dar respuesta │
 │ - Tools   │            │               │        │ devolver     │        │ final al      │
 │ - Prompt  │            │ stop_reason:  │        │ resultado    │        │ usuario       │
 │           │            │ "tool_use"    │        │              │        │               │
 └───────────┘            └───────────────┘        └──────────────┘        └───────────────┘

      TU                       CLAUDE                    TU                     CLAUDE
```

---

#### **Paso 1:** Proporcionar herramientas y prompt (API Request)

- Define las herramientas disponibles: **nombre**, **descripción** y **esquema de entrada** (input schema)
- Envía un prompt del usuario que podría requerir una o más herramientas
- *Ejemplo:* `"¿Cuántas acciones de General Motors puedo comprar con $500?"`

#### **Paso 2:** Claude usa una herramienta (API Response)

- Claude evalúa el prompt y decide si alguna herramienta es útil
- Si es así, decide **cuál(es)** usar y con **qué parámetros**
- Genera una solicitud de herramienta con formato específico
- La respuesta de la API tendrá `stop_reason: "tool_use"`

#### **Paso 3:** Extraer input, ejecutar código, devolver resultados (API Request)

- **Tu código** extrae el nombre de la herramienta y los inputs de la respuesta de Claude
- **Tu código** ejecuta la función real (ej. llamar a una API de mercado de valores)
- Devuelves los resultados a Claude en un bloque `tool_result`

#### **Paso 4:** Claude formula la respuesta final (API Response)

- Claude recibe los resultados de la herramienta
- Incorpora la información en su respuesta final al usuario

> **Nota:** Los pasos 3 y 4 son **opcionales** — en algunos flujos de trabajo, solo necesitas que Claude invoque la herramienta y no necesitas devolver resultados. Más sobre esto en lecciones posteriores.

### Paso 1: Proporcionar herramientas y prompt a Claude (API Request)

A continuación, definimos la herramienta `get_stock_price` con su **nombre**, **descripción** y **esquema de entrada** (input schema). Profundizaremos en la definición de herramientas más adelante, pero aquí hay una definición hipotética:

````python
# Definición de la herramienta con JSON Schema
tool_definition = {
    "name": "get_stock_price",                                               # Nombre único de la herramienta
    "description": "Retrieves the current stock price for a given company",  # Descripción clara
    "input_schema": {                                                        # Esquema JSON de los parámetros
        "type": "object",
        "properties": {
            "company": {
                "type": "string",
                "description": "The company name or stock ticker to fetch data for"
            }
        },
        "required": ["company"]                                             # Parámetros obligatorios
    }
}
````

Luego le informamos a Claude que esta herramienta existe y hacemos una solicitud con un prompt que podría requerir su uso:

````python
# Enviamos el prompt junto con la definición de herramientas
response = client.messages.create(
    model="claude-sonnet-4-20250514",
    messages=[
        {
            "role": "user",
            "content": "¿Cuántas acciones de General Motors puedo comprar con $500?"
        }
    ],
    max_tokens=500,
    tools=[tool_definition]  # ← Aquí le pasamos las herramientas disponibles
)
````

### Paso 0: Escribir la funcion de la calculadora

El primer paso es definir la funcion real de la calculadora y asegurarnos de que funcione, **independientemente de Claude**. Escribiremos una funcion MUY simple que espera tres argumentos:

- Una **operacion** como `"add"` o `"multiply"`
- Dos **operandos** numericos

Aqui la implementacion basica:

````python
def calculator(operation, operand1, operand2):
    if operation == "add":
        return operand1 + operand2
    elif operation == "subtract":
        return operand1 - operand2
    elif operation == "multiply":
        return operand1 * operand2
    elif operation == "divide":
        if operand2 == 0:
            raise ValueError("No se puede dividir entre cero.")
        return operand1 / operand2
    else:
        raise ValueError(f"Operacion no soportada: {operation}")
````

Ten en cuenta que esta funcion es bastante limitada porque solo puede manejar expresiones simples como `234 + 213` o `3 * 9`. El objetivo aqui es recorrer el **proceso de trabajar con herramientas** mediante un ejemplo educativo simple.

Probemos nuestra funcion para asegurarnos de que funcione:

````python
calculator("add", 10, 3)
````

````python
calculator("divide", 200, 25)
````

### Paso 1: Definir la herramienta para Claude

El siguiente paso es definir nuestra herramienta e informarle a Claude sobre ella. Al definir una herramienta, seguimos un **formato muy especifico**. Cada definicion de herramienta incluye:

| `input_schema` | Un objeto JSON Schema que define los parámetros esperados |

Aqui un ejemplo simple para una herramienta hipotetica:

```json

  "description": "Sends an email to the specified recipient with the given subject and body.",

    "type": "object",
    "properties": {

        "type": "string",
        "description": "The email address of the recipient"
      },
      "subject": {
        "type": "string",
        "description": "The subject line of the email"
      },
      "body": {
        "type": "string",
        "description": "The content of the email message"
      }
    },
    "required": ["to", "subject", "body"]
  }
}
```

Esta herramienta, llamada `send_email`, espera los siguientes inputs:

| Input | Tipo | Requerido |
|-------|------|-----------|
| `to` | string | Si |
| `subject` | string | Si |
| `body` | string | Si |

Aqui otra definicion de herramienta para una herramienta llamada `search_product`:

```json
{
  "name": "search_product",

  "input_schema": {

    "properties": {
```

### Definicion de nuestra herramienta calculadora

Definamos la herramienta correspondiente a nuestra funcion `calculator` que escribimos antes. Sabemos que la funcion tiene 3 argumentos requeridos:

| Argumento | Tipo | Restricciones |
|-----------|------|---------------|
| `operation` | string | Solo puede ser `"add"`, `"subtract"`, `"multiply"` o `"divide"` |
| `operand1` | number | Cualquier numero |
| `operand2` | number | Cualquier numero |

Aqui la definicion de la herramienta:

````python
calculator_tool = {
    "name": "calculator",
    "description": "A simple calculator that performs basic arithmetic operations.",
    "input_schema": {
        "type": "object",
        "properties": {
            "operation": {
                "type": "string",
                "enum": ["add", "subtract", "multiply", "divide"],
                "description": "The arithmetic operation to perform."
            },
            "operand1": {
                "type": "number",
                "description": "The first operand."
            },
            "operand2": {
                "type": "number",
                "description": "The second operand."
            }
        },
        "required": ["operation", "operand1", "operand2"]
    }
}
````

### Nota importante: Claude a veces es "demasiado entusiasta"

Si le preguntamos a Claude algo que **no requiere tool use** (en este caso, algo que no tiene nada que ver con matematicas), esperariamos que responda normalmente. Claude usualmente lo hace, pero a veces es **muy entusiasta** con sus herramientas.

Veamos qué pasa cuando le preguntamos "What color are emeralds?" (una pregunta que claramente no necesita calculadora):

````python
# Pregunta que NO requiere la calculadora
response = client.messages.create(
    model="claude-3-haiku-20240307",
    messages=[{"role": "user", "content": "What color are emeralds?"}],
    max_tokens=400,
    tools=[calculator_tool]
)
````

Claude quiere que llamemos a la herramienta calculadora para una pregunta sobre colores. Una solucion muy sencilla es ajustar nuestro prompt o agregar un **system prompt** que diga algo como: *"Tienes acceso a herramientas, pero solo usalas cuando sea necesario. Si no se requiere una herramienta, responde normalmente"*:

### Solucion: Ejemplo completo con system prompt

````python
def calculator(operation, operand1, operand2):
    if operation == "add":
        return operand1 + operand2
    elif operation == "subtract":
        return operand1 - operand2
    elif operation == "multiply":
        return operand1 * operand2
    elif operation == "divide":
        if operand2 == 0:
            raise ValueError("No se puede dividir entre cero.")
        return operand1 / operand2
    else:
        raise ValueError(f"Operacion no soportada: {operation}")


calculator_tool = {
    "name": "calculator",
    "description": "A simple calculator that performs basic arithmetic operations.",
    "input_schema": {
        "type": "object",
        "properties": {
            "operation": {
                "type": "string",
                "enum": ["add", "subtract", "multiply", "divide"],
                "description": "The arithmetic operation to perform.",
            },
            "operand1": {"type": "number", "description": "The first operand."},
            "operand2": {"type": "number", "description": "The second operand."},
        },
        "required": ["operation", "operand1", "operand2"],
    },
}


def prompt_claude(prompt):
    """Funcion que envia un prompt a Claude con acceso a la calculadora."""
    messages = [{"role": "user", "content": prompt}]
    response = client.messages.create(
        model="claude-3-haiku-20240307",
        # System prompt que le dice a Claude que solo use herramientas cuando sea necesario
        system="You have access to tools, but only use them when necessary. If a tool is not required, respond as normal",
        messages=messages,
        max_tokens=500,
        tools=[calculator_tool],
    )

    # Si Claude quiere usar una herramienta
    if response.stop_reason == "tool_use":
        tool_use = response.content[-1]
        tool_name = tool_use.name
        tool_input = tool_use.input

        if tool_name == "calculator":
            print("Claude quiere usar la herramienta calculadora")
            operation = tool_input["operation"]
            operand1 = tool_input["operand1"]
            operand2 = tool_input["operand2"]

            try:
                result = calculator(operation, operand1, operand2)
                print("Resultado del calculo:", result)
            except ValueError as e:
                print(f"Error: {str(e)}")

    # Si Claude responde normalmente sin herramientas
    elif response.stop_reason == "end_turn":
        print("Claude no quiso usar ninguna herramienta")
        print("Claude respondio con:")
        print(response.content[0].text)
````

## La idea clave

Una de las formas mas interesantes de utilizar tool use es **forzar a Claude a responder con contenido estructurado como JSON**. Hay muchas situaciones en las que necesitamos obtener una respuesta JSON estandarizada: extraer entidades, resumir datos, analizar sentimiento, etc.

Una forma de hacerlo es simplemente pedirle a Claude que responda con JSON, pero esto puede requerir trabajo adicional para extraer el JSON de la cadena de texto o para asegurarse de que siga el formato exacto que queremos.

La buena noticia es que cada vez que Claude quiere usar una herramienta, **ya responde usando el formato perfectamente estructurado** que le indicamos al definir la herramienta.

En la leccion anterior, le dimos a Claude una herramienta calculadora. Cuando quiso usarla, respondio con contenido como este:

```json
{
    "operand1": 1984135,
    "operand2": 9343116,
    "operation": "multiply"
}
```

Eso se parece **sospechosamente a JSON**.

Si queremos que Claude genere JSON estructurado, podemos usar esto a nuestro favor. Solo tenemos que definir una herramienta que describa la estructura JSON deseada y luego informarle a Claude sobre ella. Claude respondera creyendo que esta "llamando una herramienta", pero en realidad lo unico que nos interesa es la **respuesta estructurada** que nos da.

## Vision general del concepto

¿En que se diferencia esto de lo que hicimos en la leccion anterior? Aqui un diagrama del flujo de la leccion pasada:

### Leccion anterior: Tool use "real"

```
  Tu App                          Claude API                     Tu Codigo
  ──────                          ──────────                     ─────────
    │                                  │                              │
    │  tools[] + prompt                │                              │
    │─────────────────────────────────►│                              │
    │                                  │                              │
    │         stop_reason: tool_use    │                              │
    │         tool: calculator         │                              │
    │         input: {op, a, b}        │                              │
    │◄─────────────────────────────────│                              │
    │                                  │                              │
    │  Extraer inputs de la respuesta  │                              │
    │──────────────────────────────────────────────────────────────── │
    │                                  │                              │
    │                                  │     calculator("multiply",   │
    │                                  │                1984135,      │
    │                                  │                9343116)      │
    │                                  │              │               │
    │                                  │              ▼               │
    │                                  │    resultado = 18538003464660│
    │                                  │                              │
    │  Se EJECUTA la herramienta real                                 │
    └─────────────────────────────────────────────────────────────────┘
```

En la leccion anterior, le dimos a Claude acceso a una herramienta, Claude quiso llamarla y luego nosotros **ejecutamos realmente** la funcion subyacente.

En esta leccion, vamos a "engañar" a Claude diciendole que tiene acceso a una herramienta, pero **no necesitaremos ejecutar** la funcion subyacente. Usamos la herramienta como una forma de forzar una estructura particular de respuesta:

### Esta leccion: Tool use como "truco" para JSON

```
  Tu App                          Claude API
  ──────                          ──────────
    │                                  │
    │  tools[] + prompt                │
    │  (la herramienta define la       │
    │   FORMA del JSON que queremos)   │
    │─────────────────────────────────►│
    │                                  │
    │         stop_reason: tool_use    │
    │         tool: print_sentiment    │
    │         input: {                 │
    │           positive_score: 0.1,   │
    │           negative_score: 0.8,   │
    │           neutral_score: 0.1     │
    │         }                        │
    │◄─────────────────────────────────│
    │                                  │
    │  Extraer inputs = JSON listo     │
    │  NO se ejecuta ninguna funcion   │
    │                                  │
    │  resultado = input del tool_use  │
    │  (ya es JSON estructurado)       │
    └──────────────────────────────────┘

    DIFERENCIA CLAVE: No hay Paso 3 ni Paso 4.
    Solo nos quedamos con los datos estructurados
    que Claude genero al "intentar llamar" la herramienta.
```

## Ejemplo 1: Analisis de Sentimiento

Supongamos que queremos que Claude analice el sentimiento de un texto y responda con un objeto JSON que siga esta estructura:

```json
{
  "negative_score": 0.6,
  "neutral_score": 0.3,
  "positive_score": 0.1
}
```

Todo lo que debemos hacer es definir una herramienta que capture esta estructura usando JSON Schema:

````python
# Definimos una herramienta cuyo input_schema describe la estructura JSON que queremos
tools = [
    {
        "name": "print_sentiment_scores",
        "description": "Prints the sentiment scores of a given text.",
        "input_schema": {
            "type": "object",
            "properties": {
                "positive_score": {"type": "number", "description": "The positive sentiment score, ranging from 0.0 to 1.0."},
                "negative_score": {"type": "number", "description": "The negative sentiment score, ranging from 0.0 to 1.0."},
                "neutral_score": {"type": "number", "description": "The neutral sentiment score, ranging from 0.0 to 1.0."}
            },
            "required": ["positive_score", "negative_score", "neutral_score"]
        }
    }
]
````

Ahora podemos informarle a Claude sobre esta herramienta y **decirle explicitamente que la use**, para asegurarnos de que realmente lo haga. Deberiamos obtener una respuesta indicando que Claude quiere usar la herramienta, y esa respuesta contendra todos los datos en el formato exacto que queremos:

````python
from anthropic import Anthropic
from dotenv import load_dotenv
import json

load_dotenv()
client = Anthropic()

# Texto de ejemplo con sentimiento negativo
tweet = "I'm a HUGE hater of pickles.  I actually despise pickles.  They are garbage."

# Incluimos instruccion explicita de usar la herramienta
query = f"""
<text>
{tweet}
</text>

Only use the print_sentiment_scores tool.
"""

response = client.messages.create(
    model="claude-3-sonnet-20240229",
    max_tokens=4096,
    tools=tools,
    messages=[{"role": "user", "content": query}]
)
````

Veamos la respuesta. La parte importante esta en el campo `input` del `ToolUseBlock`:

```
ToolsBetaMessage(
  content=[
    TextBlock(text='Here is the sentiment analysis...', type='text'),
    ToolUseBlock(
      input={'positive_score': 0.0, 'negative_score': 0.791, 'neutral_score': 0.209},
      name='print_sentiment_scores',
      type='tool_use'
    )
  ],
  stop_reason='tool_use', ...
)
```

Observa: el campo `input` ya contiene **exactamente** la estructura JSON que definimos en nuestro schema.

Claude "cree" que esta llamando a una herramienta que usara estos datos de analisis de sentimiento, pero en realidad solo vamos a **extraer los datos y convertirlos en JSON**:

````python
import json

# Extraemos los datos estructurados de la respuesta de Claude
json_sentiment = None
for content in response.content:
    if content.type == "tool_use" and content.name == "print_sentiment_scores":
        json_sentiment = content.input
        break

if json_sentiment:
    print("Analisis de Sentimiento (JSON):")
    print(json.dumps(json_sentiment, indent=2))
else:
    print("No se encontro analisis de sentimiento en la respuesta.")
````

Resultado:

```json
{
  "positive_score": 0.0,
  "negative_score": 0.791,
  "neutral_score": 0.209
}
```

Funciona. Obtuvimos JSON perfectamente estructurado sin necesidad de parsear texto libre.

Ahora convirtamos eso en una **funcion reutilizable** que reciba un tweet o articulo e imprima el analisis de sentimiento como JSON:

````python
def analyze_sentiment(content):
    """Analiza el sentimiento de un texto y devuelve JSON estructurado."""

    query = f"""
    <text>
    {content}
    </text>

    Only use the print_sentiment_scores tool.
    """

    response = client.messages.create(
        model="claude-3-sonnet-20240229",
        max_tokens=4096,
        tools=tools,
        messages=[{"role": "user", "content": query}]
    )

    json_sentiment = None
    for content in response.content:
        if content.type == "tool_use" and content.name == "print_sentiment_scores":
            json_sentiment = content.input
            break

    if json_sentiment:
        print("Analisis de Sentimiento (JSON):")
        print(json.dumps(json_sentiment, indent=2))
    else:
        print("No se encontro analisis de sentimiento en la respuesta.")
````

````python
# Probemos con un texto positivo
analyze_sentiment("OMG I absolutely love taking bubble baths soooo much!!!!")

# Analisis de Sentimiento (JSON):
# {
#   "positive_score": 0.8,
#   "negative_score": 0.0,
#   "neutral_score": 0.2
# }
````

````python
# Probemos con un texto neutral
analyze_sentiment("Honestly I have no opinion on taking baths")

# Analisis de Sentimiento (JSON):
# {
#   "positive_score": 0.056,
#   "negative_score": 0.065,
#   "neutral_score": 0.879
# }
````

### Forzando tool use con `tool_choice`

Actualmente estamos "forzando" a Claude a usar nuestra herramienta `print_sentiment_scores` **mediante el prompt** (escribimos *"Only use the print_sentiment_scores tool"*). Esto usualmente funciona, pero hay una forma mejor.

Podemos forzar a Claude a usar una herramienta especifica mediante el parametro **`tool_choice`**:

````python
# Esto le dice a Claude que DEBE responder llamando a print_sentiment_scores
tool_choice = {"type": "tool", "name": "print_sentiment_scores"}
````

El codigo anterior le indica a Claude que **debe** responder llamando a la herramienta `print_sentiment_scores`. Actualicemos nuestra funcion para usarlo:

````python
def analyze_sentiment(content):
    """Analiza sentimiento forzando el uso de la herramienta via tool_choice."""

    query = f"""
    <text>
    {content}
    </text>
    """

    response = client.messages.create(
        model="claude-3-sonnet-20240229",
        max_tokens=4096,
        tools=tools,
        # Forzamos a Claude a usar esta herramienta especifica
        tool_choice={"type": "tool", "name": "print_sentiment_scores"},
        messages=[{"role": "user", "content": query}]
    )

    json_sentiment = None
    for content in response.content:
        if content.type == "tool_use" and content.name == "print_sentiment_scores":
            json_sentiment = content.input
            break

    if json_sentiment:
        print("Analisis de Sentimiento (JSON):")
        print(json.dumps(json_sentiment, indent=2))
    else:
        print("No se encontro analisis de sentimiento en la respuesta.")
````

Cubriremos `tool_choice` con mayor detalle en una leccion posterior.

````python
import wikipedia

# Definicion de la herramienta con la estructura compleja que queremos
tools = [
    {
        "name": "print_article_classification",
        "description": "Prints the classification results.",
        "input_schema": {
            "type": "object",
            "properties": {
                "subject": {
                    "type": "string",
                    "description": "The overall subject of the article",
                },
                "summary": {
                    "type": "string",
                    "description": "A paragaph summary of the article"
                },
                "keywords": {
                    "type": "array",
                    "items": {
                        "type": "string",
                        "description": "List of keywords and topics in the article"
                    }
                },
                "categories": {
                    "type": "array",
                    "items": {
                        "type": "object",
                        "properties": {
                            "name": {"type": "string", "description": "The category name."},
                            "score": {"type": "number", "description": "The classification score for the category, ranging from 0.0 to 1.0."}
                        },
                        "required": ["name", "score"]
                    }
                }
            },
            "required": ["subject", "summary", "keywords", "categories"]
        }
    }
]


def generate_json_for_article(subject):
    """Genera una clasificacion JSON estructurada para un articulo de Wikipedia."""
    page = wikipedia.page(subject, auto_suggest=True)
    query = f"""
    <document>
    {page.content}
    </document>

    Use the print_article_classification tool. Example categories are Politics, Sports, Technology, Entertainment, Business.
    """

    response = client.messages.create(
        model="claude-3-haiku-20240307",
        max_tokens=4096,
        tools=tools,
        messages=[{"role": "user", "content": query}]
    )

    json_classification = None
    for content in response.content:
        if content.type == "tool_use" and content.name == "print_article_classification":
            json_classification = content.input
            break

    if json_classification:
        print("Clasificacion de Texto (JSON):")
        print(json.dumps(json_classification, indent=2))
    else:
        print("No se encontro clasificacion en la respuesta.")
````

## Construyendo el Flujo Completo

### 1. Definir nuestra funcion de busqueda en Wikipedia

Antes de hacer nada con Claude, escribamos una funcion simple para buscar en Wikipedia. La siguiente funcion usa el paquete `wikipedia` para buscar paginas que coincidan con un termino de busqueda. Para simplificar, tomamos el primer titulo devuelto y accedemos al contenido de esa pagina.

> **Nota:** Esta funcion simple asume que siempre encontrara un articulo de Wikipedia. Para mantenerla breve, no incluye manejo de errores, lo cual no es recomendable en un entorno de produccion.

````python
import wikipedia

def get_article(search_term):
    """Busca un articulo de Wikipedia y devuelve su contenido."""
    results = wikipedia.search(search_term)
    first_result = results[0]
    page = wikipedia.page(first_result, auto_suggest=False)
    return page.content
````

### 2. Escribir la definicion de la herramienta

Ahora necesitamos definir nuestra herramienta usando el formato **JSON Schema** adecuado. Esta es una definicion muy simple porque la funcion espera un solo argumento: el termino de busqueda (string).

````python
# Definicion de la herramienta de busqueda en Wikipedia
article_search_tool = {
    "name": "get_article",
    "description": "A tool to retrieve an up to date Wikipedia article.",
    "input_schema": {
        "type": "object",
        "properties": {
            "search_term": {
                "type": "string",
                "description": "The search term to find a wikipedia article by title"
            },
        },
        "required": ["search_term"]
    }
}
````

### 3. Enviar a Claude la herramienta y el prompt del usuario

Ahora le indicamos a Claude que tiene acceso a la herramienta de busqueda en Wikipedia y le hacemos una pregunta que sabemos que **no puede responder sin la herramienta**, como "Who won the 2024 Masters Tournament?"

````python
# Paso 1: Crear el cliente y enviar el mensaje con la herramienta
from anthropic import Anthropic
from dotenv import load_dotenv

load_dotenv()

client = Anthropic()

messages = [{"role": "user", "content": "who won the 2024 Masters Tournament?"}]

response = client.messages.create(
    model="claude-3-sonnet-20240229",
    messages=messages,
    max_tokens=1000,
    tools=[article_search_tool]  # Le damos acceso a nuestra herramienta
)
````

#### Formato del mensaje `tool_result`

Ahora que ejecutamos la funcion, debemos **responder a Claude** con los datos de Wikipedia. Recordemos que cuando Claude quiere usar una herramienta, responde con:

| Campo | Descripcion |
|-------|-------------|
| `id` | Identificador unico para este `ToolUseBlock`. Se usa para vincular con el resultado. |
| `name` | Nombre de la herramienta que quiere usar |
| `input` | Objeto con los argumentos, conforme al `input_schema` |

Para continuar la conversacion, necesitamos enviar un nuevo mensaje con `role: "user"` y un bloque de contenido tipo `tool_result` que incluya:

| Campo | Descripcion | Requerido |
|-------|-------------|-----------|
| `tool_use_id` | El `id` del `ToolUseBlock` al cual responde este resultado | Si |
| `content` | El resultado de la herramienta (string o lista de bloques) | Si |
| `is_error` | `true` si la ejecucion resulto en error | No (opcional) |

Aqui un ejemplo de como se ve un mensaje `tool_result` correctamente formateado:

````python
# Estructura de un mensaje tool_result (ejemplo)
{
  "role": "user",
  "content": [
    {
      "type": "tool_result",
      "tool_use_id": "toolu_01A09q90qw90lq917835lq9",  # DEBE coincidir con el id del ToolUseBlock
      "content": "El resultado de ejecutar la herramienta va aqui"
    }
  ]
}
````

Ahora hagamoslo para nuestro ejemplo de Wikipedia. Necesitamos construir un mensaje de respuesta de herramienta correctamente formateado para enviar el resultado de la busqueda a Claude:

````python
# Construir el mensaje tool_result con el contenido de Wikipedia
tool_response = {
    "role": "user",
    "content": [
        {
        "type": "tool_result",
        "tool_use_id": tool_use.id,  # Usamos el id del ToolUseBlock de Claude
        "content": wiki_result        # El contenido del articulo de Wikipedia
        }
    ]
}
````

## Mejorando el Codigo

Como minimo, deberiamos poner todo el codigo anterior en una **funcion reutilizable** para poder probarlo varias veces:

````python
def answer_question(question):
    """Funcion que implementa el flujo completo de tool use con Wikipedia."""
    messages = [{"role": "user", "content": question}]

    # Paso 1: Enviar la pregunta con la herramienta disponible
    response = client.messages.create(
        model="claude-3-sonnet-20240229",
        messages=messages,
        max_tokens=1000,
        tools=[article_search_tool]
    )
    
    # Verificar si Claude quiere usar una herramienta
    if(response.stop_reason == "tool_use"):
        tool_use = response.content[-1]
        tool_name = tool_use.name
        tool_input = tool_use.input
        # Paso 2: Agregar la respuesta de Claude a messages
        messages.append({"role": "assistant", "content": response.content})

        if tool_name == "get_article":
            search_term = tool_input["search_term"]
            print(f"Claude quiere obtener un articulo sobre: {search_term}")
            # Paso 3: Ejecutar la herramienta real
            wiki_result = get_article(search_term)
            # Construir el mensaje tool_result
            tool_response = {
                "role": "user",
                "content": [
                    {
                    "type": "tool_result",
                    "tool_use_id": tool_use.id,
                    "content": wiki_result
                    }
                ]
            }
            messages.append(tool_response)
            # Paso 4: Enviar todo de vuelta a Claude para la respuesta final
            response = client.messages.create(
                model="claude-3-sonnet-20240229",
                messages=messages,
                max_tokens=1000,
                tools=[article_search_tool]
            )
            print("Respuesta final de Claude:")
            print(response.content[0].text)

    else:
        # Claude no necesito la herramienta
        print("Claude no uso nuestra herramienta")
        print(response.content[0].text)
````

Podriamos agregar un system prompt como el siguiente:

````python
def answer_question(question):
    """Funcion mejorada con system prompt para controlar el uso de herramientas."""
    system_prompt = """
    You will be asked a question by the user. 
    If answering the question requires data you were not trained on, you can use the get_article tool to get the contents of a recent wikipedia article about the topic. 
    If you can answer the question without needing to get more information, please do so. 
    Only call the tool when needed. 
    """

    messages = [{"role": "user", "content": question}]

    # Paso 1: Enviar con system prompt + herramienta
    response = client.messages.create(
        model="claude-3-sonnet-20240229",
        system=system_prompt,  # Ahora incluimos el system prompt
        messages=messages,
        max_tokens=1000,
        tools=[article_search_tool]
    )
    
    if(response.stop_reason == "tool_use"):
        tool_use = response.content[-1]
        tool_name = tool_use.name
        tool_input = tool_use.input
        messages.append({"role": "assistant", "content": response.content})

        if tool_name == "get_article":
            search_term = tool_input["search_term"]
            print(f"Claude quiere obtener un articulo sobre: {search_term}")
            wiki_result = get_article(search_term)
            tool_response = {
                "role": "user",
                "content": [
                    {
                    "type": "tool_result",
                    "tool_use_id": tool_use.id,
                    "content": wiki_result
                    }
                ]
            }
            messages.append(tool_response)
            # Paso 4 con system prompt incluido
            response = client.messages.create(
                model="claude-3-sonnet-20240229",
                system=system_prompt,  # Incluir system prompt en cada llamada
                messages=messages,
                max_tokens=1000,
                tools=[article_search_tool]
            )
            print("Respuesta final de Claude:")
            print(response.content[0].text)

    else:
        print("Claude no uso nuestra herramienta")
        print(response.content[0].text)
````

# Tool Choice (Seleccion de Herramientas)

## Auto

Establecer `tool_choice` en `auto` permite que el modelo **decida automaticamente** si usar herramientas o no. Este es el **comportamiento por defecto** cuando trabajas con herramientas sin especificar `tool_choice`.

Para demostrarlo, vamos a proporcionar a Claude una herramienta de busqueda web ficticia. Le haremos preguntas: algunas que requieran la herramienta y otras que pueda responder por si solo.

Definimos una herramienta llamada `web_search`. **Nota:** para mantener esta demo simple, no estamos buscando en la web realmente.

````python
# Funcion simulada de busqueda web (no busca realmente)
def web_search(topic):
    print(f"Simulando busqueda web sobre: {topic}")

# Definicion de la herramienta
web_search_tool = {
    "name": "web_search",
    "description": "A tool to retrieve up to date information on a given topic by searching the web",
    "input_schema": {
        "type": "object",
        "properties": {
            "topic": {
                "type": "string",
                "description": "The topic to search the web for"
            },
        },
        "required": ["topic"]
    }
}
````

Ahora escribimos una funcion que acepta una consulta del usuario, la pasa a Claude junto con `web_search_tool`, y establece `tool_choice` en `auto`:

```python
tool_choice={"type": "auto"}  # Claude decide si usar la herramienta o no
```

Aqui la funcion completa:

````python
from datetime import date

def chat_with_web_search(user_query):
    messages = [{"role": "user", "content": user_query}]

    system_prompt = f"""
    Answer as many questions as you can using your existing knowledge.  
    Only search the web for queries that you can not confidently answer.
    Today's date is {date.today().strftime("%B %d %Y")}
    If you think a user's question involves something in the future that hasn't happened yet, use the search tool.
    """

    response = client.messages.create(
        system=system_prompt,
        model="claude-3-sonnet-20240229",
        messages=messages,
        max_tokens=1000,
        tool_choice={"type": "auto"},  # Claude decide libremente
        tools=[web_search_tool]
    )

    last_content_block = response.content[-1]
    if last_content_block.type == "text":
        print("Claude NO llamo a una herramienta")
        print(f"Asistente: {last_content_block.text}")
    elif last_content_block.type == "tool_use":
        print("Claude quiere usar una herramienta")
        print(last_content_block)
````

La idea es que Claude **siempre llame a alguna herramienta** y nunca responda directamente con texto. Para asegurar esto, establecemos `tool_choice` en `any`:

````python
# Claude DEBE usar alguna herramienta, pero elige cual
tool_choice={"type": "any"}
````

````python
# Funciones reales para las herramientas del chatbot SMS
def send_text_to_user(text):
    """Envia un mensaje de texto al usuario (simulado con print)."""
    print(f"MENSAJE DE TEXTO ENVIADO: {text}")

def get_customer_info(username):
    """Obtiene informacion del cliente (datos simulados)."""
    return {
        "username": username,
        "email": f"{username}@email.com",
        "purchases": [
            {"id": 1, "product": "computer mouse"},
            {"id": 2, "product": "screen protector"},
            {"id": 3, "product": "usb charging cable"},
        ]
    }

# Definicion de las herramientas
tools = [
    {
        "name": "send_text_to_user",
        "description": "Sends a text message to a user",
        "input_schema": {
            "type": "object",
            "properties": {
                "text": {"type": "string", "description": "The piece of text to be sent to the user via text message"},
            },
            "required": ["text"]
        }
    },
    {
        "name": "get_customer_info",
        "description": "gets information on a customer based on the customer's username.  Response includes email, username, and previous purchases. Only call this tool once a user has provided you with their username",
        "input_schema": {
            "type": "object",
            "properties": {
                "username": {"type": "string", "description": "The username of the user in question. "},
            },
            "required": ["username"]
        }
    },
]

# System prompt para el chatbot SMS
system_prompt = """
All your communication with a user is done via text message.
Only call tools when you have enough information to accurately call them.  
Do not call the get_customer_info tool until a user has provided you with their username. This is important.
If you do not know a user's username, simply ask a user for their username.
"""

def sms_chatbot(user_message):
    """Chatbot SMS que SIEMPRE responde mediante una herramienta."""
    messages = [{"role": "user", "content": user_message}]

    response = client.messages.create(
        system=system_prompt,
        model="claude-3-sonnet-20240229",
        max_tokens=4096,
        tools=tools,
        tool_choice={"type": "any"},  # DEBE usar alguna herramienta
        messages=messages
    )

    if response.stop_reason == "tool_use":
        last_content_block = response.content[-1]
        if last_content_block.type == 'tool_use':
            tool_name = last_content_block.name
            tool_inputs = last_content_block.input
            print(f"======= Claude quiere llamar a: {tool_name} =======")
            if tool_name == "send_text_to_user":
                send_text_to_user(tool_inputs["text"])
            elif tool_name == "get_customer_info":
                print(get_customer_info(tool_inputs["username"]))
            else:
                print("Esa herramienta no existe!")
    else:
        print("No se llamo ninguna herramienta. Esto no deberia pasar!")
````

### Puntos Clave de esta Seccion

| Concepto | Detalle |
|----------|---------|
| **`tool_choice: auto`** | Claude decide si usar herramientas o no. Es el valor por defecto. Requiere un buen system prompt. |
| **`tool_choice: any`** | Claude **debe** usar alguna herramienta, pero elige cual. Ideal para chatbots donde toda comunicacion pasa por tools. |
| **`tool_choice: tool`** | Claude **siempre** usa la herramienta especificada. Ideal para forzar JSON estructurado o extraccion de datos. |
| **System prompt** | Fundamental con `auto` para evitar que Claude sea demasiado entusiasta al llamar herramientas. |
| **Nombre requerido** | Con `type: "tool"`, debes especificar `"name"` de la herramienta a forzar. |

<details>
<summary>Cuando usar cada opcion</summary>

- **auto:** Cuando quieres flexibilidad. Claude decide basandose en el contexto.
- **any:** Cuando toda interaccion DEBE pasar por una herramienta (ej: chatbot SMS, sistemas donde no hay respuesta directa).
- **tool:** Cuando necesitas garantizar que se llame a una herramienta especifica cada vez (ej: extraccion de datos, formato JSON).

</details>

## Escribiendo Nuestras Herramientas

Como siempre, incluimos `name`, `description` y un `input_schema`. En este caso hay un unico input (`order_id`) y es requerido.

### Herramienta con `enum`: `get_user`

Ahora veamos una herramienta mas compleja que corresponde al metodo `get_user`. Recordemos que este metodo tiene dos argumentos:

| Argumento | Descripcion | Valores posibles |
|-----------|-------------|------------------|
| `key` | Atributo por el cual buscar | `"email"`, `"username"`, `"phone"` |
| `value` | El termino de busqueda | El email, telefono o username real |

Aqui esta la definicion de la herramienta:

````python
tool2 = {
    "name": "get_user",
    "description": "Looks up a user by email, phone, or username.",
    "input_schema": {
        "type": "object",
        "properties": {
            "key": {
                "type": "string",
                "enum": ["email", "phone", "username"],  # <-- enum restringe los valores validos
                "description": "The attribute to search for a user by (email, phone, or username)."
            },
            "value": {
                "type": "string",
                "description": "The value to match for the specified attribute."
            }
        },
        "required": ["key", "value"]
    }
}
````

> **Punto clave:** Observa como definimos el conjunto de opciones validas para `key` usando **`enum`** en el esquema. Esto le indica a Claude exactamente cuales son los valores permitidos para ese parametro.

## Dando Nuestras Herramientas a Claude

Ahora necesitamos hacer varias cosas:

1. **Informar** a Claude sobre nuestras herramientas y enviar el mensaje del usuario
2. **Manejar** la respuesta de Claude:
   - Si Claude **no** quiere usar una herramienta: imprimir la respuesta al usuario y pedir el siguiente mensaje
   - Si Claude **si** quiere usar una herramienta:
     - Verificar que Claude llamo una herramienta valida
     - Ejecutar la funcion subyacente (`get_user`, `cancel_order`, etc.)
     - Enviar a Claude el resultado de la herramienta

Eventualmente el objetivo es crear un **script interactivo** con un chatbot en bucle. Pero para mantener las cosas simples, empezaremos escribiendo el codigo de forma lineal.

### Funcion puente: `process_tool_call`

Primero escribimos una funcion que traduce la respuesta de tool use de Claude a una llamada real a nuestra base de datos:

````python
def process_tool_call(tool_name, tool_input):
    """Ejecuta la herramienta correspondiente en nuestra base de datos."""
    if tool_name == "get_user":
        return db.get_user(tool_input["key"], tool_input["value"])
    elif tool_name == "get_order_by_id":
        return db.get_order_by_id(tool_input["order_id"])
    elif tool_name == "get_customer_orders":
        return db.get_customer_orders(tool_input["customer_id"])
    elif tool_name == "cancel_order":
        return db.cancel_order(tool_input["order_id"])
````

````python
import anthropic
import json

client = anthropic.Client()
MODEL_NAME = "claude-3-sonnet-20240229"

# Recreamos la base de datos y las funciones necesarias
class FakeDatabase:
    def __init__(self):
        self.customers = [
            {"id": "1213210", "name": "John Doe", "email": "john@gmail.com", "phone": "123-456-7890", "username": "johndoe"},
            {"id": "2837622", "name": "Priya Patel", "email": "priya@candy.com", "phone": "987-654-3210", "username": "priya123"},
            {"id": "3924156", "name": "Liam Nguyen", "email": "lnguyen@yahoo.com", "phone": "555-123-4567", "username": "liamn"},
            {"id": "4782901", "name": "Aaliyah Davis", "email": "aaliyahd@hotmail.com", "phone": "111-222-3333", "username": "adavis"},
            {"id": "5190753", "name": "Hiroshi Nakamura", "email": "hiroshi@gmail.com", "phone": "444-555-6666", "username": "hiroshin"},
            {"id": "6824095", "name": "Fatima Ahmed", "email": "fatimaa@outlook.com", "phone": "777-888-9999", "username": "fatimaahmed"},
            {"id": "7135680", "name": "Alejandro Rodriguez", "email": "arodriguez@protonmail.com", "phone": "222-333-4444", "username": "alexr"},
            {"id": "8259147", "name": "Megan Anderson", "email": "megana@gmail.com", "phone": "666-777-8888", "username": "manderson"},
            {"id": "9603481", "name": "Kwame Osei", "email": "kwameo@yahoo.com", "phone": "999-000-1111", "username": "kwameo"},
            {"id": "1057426", "name": "Mei Lin", "email": "meilin@gmail.com", "phone": "333-444-5555", "username": "mlin"}
        ]
        self.orders = [
            {"id": "24601", "customer_id": "1213210", "product": "Wireless Headphones", "quantity": 1, "price": 79.99, "status": "Shipped"},
            {"id": "13579", "customer_id": "1213210", "product": "Smartphone Case", "quantity": 2, "price": 19.99, "status": "Processing"},
            {"id": "97531", "customer_id": "2837622", "product": "Bluetooth Speaker", "quantity": 1, "price": "49.99", "status": "Shipped"},
            {"id": "86420", "customer_id": "3924156", "product": "Fitness Tracker", "quantity": 1, "price": 129.99, "status": "Delivered"},
            {"id": "54321", "customer_id": "4782901", "product": "Laptop Sleeve", "quantity": 3, "price": 24.99, "status": "Shipped"},
            {"id": "19283", "customer_id": "5190753", "product": "Wireless Mouse", "quantity": 1, "price": 34.99, "status": "Processing"},
            {"id": "74651", "customer_id": "6824095", "product": "Gaming Keyboard", "quantity": 1, "price": 89.99, "status": "Delivered"},
            {"id": "30298", "customer_id": "7135680", "product": "Portable Charger", "quantity": 2, "price": 29.99, "status": "Shipped"},
            {"id": "47652", "customer_id": "8259147", "product": "Smartwatch", "quantity": 1, "price": 199.99, "status": "Processing"},
            {"id": "61984", "customer_id": "9603481", "product": "Noise-Cancelling Headphones", "quantity": 1, "price": 149.99, "status": "Shipped"},
            {"id": "58243", "customer_id": "1057426", "product": "Wireless Earbuds", "quantity": 2, "price": 99.99, "status": "Delivered"},
            {"id": "90357", "customer_id": "1213210", "product": "Smartphone Case", "quantity": 1, "price": 19.99, "status": "Shipped"},
            {"id": "28164", "customer_id": "2837622", "product": "Wireless Headphones", "quantity": 2, "price": 79.99, "status": "Processing"}
        ]

    def get_user(self, key, value):
        if key in {"email", "phone", "username"}:
            for customer in self.customers:
                if customer[key] == value:
                    return customer
            return f"Couldn't find a user with {key} of {value}"
        else:
            raise ValueError(f"Invalid key: {key}")
        return None

    def get_order_by_id(self, order_id):
        for order in self.orders:
            if order["id"] == order_id:
                return order
        return None

    def get_customer_orders(self, customer_id):
        return [order for order in self.orders if order["customer_id"] == customer_id]

    def cancel_order(self, order_id):
        order = self.get_order_by_id(order_id)
        if order:
            if order["status"] == "Processing":
                order["status"] = "Cancelled"
                return "Cancelled the order"
            else:
                return "Order has already shipped.  Can't cancel it."
        return "Can't find that order!"

# Herramientas (mismo array de antes)
tools = [
    {
        "name": "get_user",
        "description": "Looks up a user by email, phone, or username.",
        "input_schema": {
            "type": "object",
            "properties": {
                "key": {"type": "string", "enum": ["email", "phone", "username"], "description": "The attribute to search for a user by (email, phone, or username)."},
                "value": {"type": "string", "description": "The value to match for the specified attribute."}
            },
            "required": ["key", "value"]
        }
    },
    {
        "name": "get_order_by_id",
        "description": "Retrieves the details of a specific order based on the order ID. Returns the order ID, product name, quantity, price, and order status.",
        "input_schema": {
            "type": "object",
            "properties": {
                "order_id": {"type": "string", "description": "The unique identifier for the order."}
            },
            "required": ["order_id"]
        }
    },
    {
        "name": "get_customer_orders",
        "description": "Retrieves the list of orders belonging to a user based on a user's customer id.",
        "input_schema": {
            "type": "object",
            "properties": {
                "customer_id": {"type": "string", "description": "The customer_id belonging to the user"}
            },
            "required": ["customer_id"]
        }
    },
    {
        "name": "cancel_order",
        "description": "Cancels an order based on a provided order_id. Only orders that are 'processing' can be cancelled",
        "input_schema": {
            "type": "object",
            "properties": {
                "order_id": {"type": "string", "description": "The order_id pertaining to a particular order"}
            },
            "required": ["order_id"]
        }
    }
]

db = FakeDatabase()

def process_tool_call(tool_name, tool_input):
    if tool_name == "get_user":
        return db.get_user(tool_input["key"], tool_input["value"])
    elif tool_name == "get_order_by_id":
        return db.get_order_by_id(tool_input["order_id"])
    elif tool_name == "get_customer_orders":
        return db.get_customer_orders(tool_input["customer_id"])
    elif tool_name == "cancel_order":
        return db.cancel_order(tool_input["order_id"])

def simple_chat():
    """Chatbot interactivo basico para TechNova."""
    user_message = input("\nUser: ")
    messages = [{"role": "user", "content": user_message}]
    while True:
        # Si el ultimo mensaje es del asistente, pedimos otro input al usuario
        if messages[-1].get("role") == "assistant":
            user_message = input("\nUser: ")
            messages.append({"role": "user", "content": user_message})

        # Enviamos solicitud a Claude
        response = client.messages.create(
            model=MODEL_NAME,
            max_tokens=4096,
            tools=tools,
            messages=messages
        )
        # Agregamos la respuesta de Claude a messages
        messages.append(
            {"role": "assistant", "content": response.content}
        )

        # Si Claude quiere usar una herramienta:
        if response.stop_reason == "tool_use":
            tool_use = response.content[-1]
            tool_name = tool_use.name
            tool_input = tool_use.input
            print(f"======Claude quiere usar la herramienta {tool_name}======")

            # Ejecutamos la herramienta en nuestra db
            tool_result = process_tool_call(tool_name, tool_input)

            # Agregamos el tool_result a messages
            messages.append(
                {
                    "role": "user",
                    "content": [
                        {
                            "type": "tool_result",
                            "tool_use_id": tool_use.id,
                            "content": str(tool_result),
                        }
                    ],
                },
            )
        else:
            # Si Claude NO quiere usar herramienta, imprimimos la respuesta
            print("\nTechNova Support: " + f"{response.content[0].text}")

# Descomenta para iniciar el chat:
# simple_chat()
````

## Problema Especifico de Opus

Cuando trabajamos con **Opus** y herramientas, el modelo frecuentemente genera sus pensamientos en tags `<thinking>` o `<reflection>` **antes** de responder al usuario. Ejemplo:

```
TechNova Support: <thinking>The user wants to cancel an order. I need 
to first look up their account. Let me ask for their information so I 
can use the get_user tool correctly.</thinking>

I'd be happy to help you cancel an order! Could you please provide me 
with your email, phone number, or username?
```

Este proceso de razonamiento generalmente lleva a **mejores resultados**, pero obviamente es una **pesima experiencia de usuario**. No queremos exponer esa logica de pensamiento al usuario.

### Solucion: Tags `<reply>` para respuestas al usuario

La solucion mas simple es pedirle al modelo que genere su respuesta orientada al usuario dentro de tags XML especificos.

### Version 3: Con tags de respuesta

````python
system_prompt = """
You are a customer support chat bot for an online retailer called TechNova. 
Your job is to help users look up their account, orders, and cancel orders.
Be helpful and brief in your responses.
You have access to a set of tools, but only use them when needed.  
If you do not have enough information to use a tool correctly, ask a user follow up questions to get the required inputs.
Do not call any of the tools unless you have the required data from a user. 

In each conversational turn, you will begin by thinking about your response. 
Once you're done, you will write a user-facing response. 
It's important to place all user-facing conversational responses in <reply></reply> XML tags to make them easy to parse.
"""
````

Las lineas clave nuevas:
- *"you will begin by thinking about your response"* -- le da permiso de razonar
- *"place all user-facing responses in `<reply></reply>` XML tags"* -- separa pensamiento de respuesta

Ahora la salida de Claude se ve asi:

```
The user wants to cancel an order. I should ask for their information 
to look up their account first.

<reply>I'd be happy to help you cancel an order! Could you please provide 
me with your email, phone number, or username?</reply>
```

Solo necesitamos **extraer** el contenido entre los tags `<reply>` y asegurarnos de que sea lo unico que se imprima al usuario.

### Version final del script con `extract_reply`

````python
import re

def extract_reply(text):
    """Extrae el contenido entre tags <reply></reply>."""
    pattern = r'<reply>(.*?)</reply>'
    match = re.search(pattern, text, re.DOTALL)
    if match:
        return match.group(1)
    else:
        return None

def simple_chat():
    """Chatbot interactivo final con system prompt y extraccion de reply."""
    system_prompt = """
    You are a customer support chat bot for an online retailer called TechNova. 
    Your job is to help users look up their account, orders, and cancel orders.
    Be helpful and brief in your responses.
    You have access to a set of tools, but only use them when needed.  
    If you do not have enough information to use a tool correctly, ask a user follow up questions to get the required inputs.
    Do not call any of the tools unless you have the required data from a user. 

    In each conversational turn, you will begin by thinking about your response. 
    Once you're done, you will write a user-facing response. 
    It's important to place all user-facing conversational responses in <reply></reply> XML tags to make them easy to parse.
    """
    user_message = input("\nUser: ")
    messages = [{"role": "user", "content": user_message}]
    while True:
        # Si el ultimo mensaje es del asistente, pedimos otro input
        if messages[-1].get("role") == "assistant":
            user_message = input("\nUser: ")
            messages.append({"role": "user", "content": user_message})

        # Enviamos solicitud a Claude (ahora con system prompt)
        response = client.messages.create(
            model=MODEL_NAME,
            system=system_prompt,
            max_tokens=4096,
            tools=tools,
            messages=messages
        )
        # Agregamos respuesta de Claude a messages
        messages.append(
            {"role": "assistant", "content": response.content}
        )

        # Si Claude quiere usar una herramienta:
        if response.stop_reason == "tool_use":
            tool_use = response.content[-1]
            tool_name = tool_use.name
            tool_input = tool_use.input
            print(f"======Claude quiere usar la herramienta {tool_name}======")

            # Ejecutamos la herramienta
            tool_result = process_tool_call(tool_name, tool_input)

            # Agregamos el tool_result
            messages.append(
                {
                    "role": "user",
                    "content": [
                        {
                            "type": "tool_result",
                            "tool_use_id": tool_use.id,
                            "content": str(tool_result),
                        }
                    ],
                },
            )
        else:
            # Extraemos solo el contenido dentro de <reply> tags
            model_reply = extract_reply(response.content[0].text)
            print("\nTechNova Support: " + f"{model_reply}")

# Descomenta para iniciar el chat:
# simple_chat()
````
